```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x  = np.linspace(0,10,500)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x**2
y4 = np.exp(-0.5*x + 5)
```

```python

fig1 = plt.figure(1)
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.subplot(2,2,4)
plt.plot(x,y4)
```

```python
fig2 = plt.figure(2)
#ax1 = fig2.add_axis([left,bottom, width, height])
sub1 = plt.subplot(2,1,1)
plt.plot(x,y3)
sub1.set(title='x**2', xlabel='x', ylabel='x*x', yticks=[], xticks =np.arange(0,11,1))
sub2 = plt.subplot(2,1,2)
plt.plot(x,y1,x,y2)
sub2.legend(['sin','cos'])
sub2.set(xlabel='x')
```

```python
fig1.savefig('4plots', dpi=1200)
fig1.savefig('2plots', dpi=100)
```
